```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# Draw scatter plot of breams and smelts

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("length")
plt.ylabel("weight")

plt.show()
```

```python
# Ready for K-Nearest Neighbors algorithm

length = bream_length + smelt_length
weight = bream_weight + smelt_weight

print(length)
print(weight)
```

```python
# The package scikit-learn requires two-dimensional list
# that hangs down each feature of the data

fish_data = [[l, w] for l, w in zip(length, weight)]
print(fish_data)
```

```python
# The first 35 fish are bream and the following 14 fish are smelt.
# Suppose that bream is "1" and smelt is "0"

fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
# fit() method of KNeighborClassifier() trains with the given data

kn.fit(fish_data, fish_target)
```

```python
# Evalulate accuracy of the model that I had created just ago
# 1 stands for 100% accuracy and 0 stands for 0%.

kn.score(fish_data, fish_target)
```

```python
# Is this a bream or a smelt? 
# predict() method of KNeighborClassifier() will predict with the real given data and the trained model.

# kn.predict([[30, 600]])

if kn.predict([[30, 600]]) == [1]:
  print("Bream")
else:
  print("Smelt")
```

```python
# Every data I had relayed to the KNeighborClassifier

print(kn._fit_X)
```

```python
print(kn._y)
```

```python
# Determine how much data will be referred during the prediction by the model

kn49 = KNeighborsClassifier(n_neighbors = 49)
```

```python
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
if kn49.score(fish_data, fish_target) == 35/49:
  print("True")
```

```python
# accuracy testing

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  kn.n_neighbors = n
  score = kn.score(fish_data, fish_target)

  print(f"n_neighbors : {n:02d} | accuracy : {score}")
```
