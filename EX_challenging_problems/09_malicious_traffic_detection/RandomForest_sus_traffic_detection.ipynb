```python
import pandas as pd
```

```python
normal_traffic_dataset  = pd.read_csv('https://raw.githubusercontent.com/KnightChaser/ML-challenge/main/EX_challenging_problems/09_malicious_traffic_detection/normal_traffic_list.csv')
anomaly_traffic_dataset = pd.read_csv('https://raw.githubusercontent.com/KnightChaser/ML-challenge/main/EX_challenging_problems/09_malicious_traffic_detection/anomaly_traffic_list.csv')
```

```python
normal_traffic_dataset['label'] = 0
normal_traffic_dataset.describe()
```

```python
anomaly_traffic_dataset['label'] = 1
anomaly_traffic_dataset.describe()
```

```python
normal_traffic_dataset = normal_traffic_dataset.dropna(axis = 0)
```

```python
anomaly_traffic_dataset = anomaly_traffic_dataset.dropna(axis = 0)
```

```python
dataset = pd.concat([normal_traffic_dataset, anomaly_traffic_dataset])
data = dataset[['entropy', 'num_pkts_out', 'num_pkts_in', 'src_port', 'total_entropy', 'duration']].to_numpy()
target = dataset['label'].to_numpy()
```

```python
print(f'data.shape: {data.shape}')
print(f'target.shape: {target.shape}')
```

```python
# generate training sets and data sets

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.2, random_state = 0xCAFE)
```

```python
# Standardization preprocessing
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled  = ss.transform(test_input)
```

```python
import matplotlib.pyplot as plt

LABELS = ["Normal", "Abnormal"]

count_classes = pd.value_counts(dataset['label'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.title('Data traffic normal/abnormal QTY comparison')
plt.xticks(range(2), LABELS)
plt.xlabel('classification')
plt.ylabel('QTY')
```

```python
# Cross-validate with random forest classification algorithm
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs = -1, random_state = 0xCAFE)
rf.fit(train_input, train_target)
```

```python
rf = RandomForestClassifier(oob_score = True, n_jobs = -1, random_state = 0xCAFE)
rf.fit(train_scaled, train_target)
print(f'train accuracy: {rf.oob_score_ * 100}%')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

pred_lr = rf.predict(test_scaled)

print(f'expectation accuracy: {accuracy_score(test_target, pred_lr) * 100}%')
print(confusion_matrix(test_target, pred_lr))
```
