The given stock CSV data(`train.csv`) has multiple items. This code snippes extracts only one items.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/KnightChaser/ML-challenge/main/EX_challenging_problems/06_stock_price_expectation/train.csv")
```

```python
df
```

```python
from tqdm import trange

item_name = input("Enter your item name: ")

df_day_list = []

index = 0
for _sequence in trange(0, 987999 + 1):
    if item_name != df.iloc[_sequence]["종목명"]:
        continue
    else:
        df_day = pd.DataFrame([[
            df.iloc[_sequence]["일자"],
            df.iloc[_sequence]["거래량"],
            df.iloc[_sequence]["시가"],
            df.iloc[_sequence]["고가"],
            df.iloc[_sequence]["저가"],
            df.iloc[_sequence]["종가"]
        ]],
        columns = ["Date", "Volume", "Open", "High", "Low", "Close"],
        index = [index])

        index += 1
        df_day_list.append(df_day)

specific_item_df = pd.concat([df for df in df_day_list], axis = 0, ignore_index=True)
```

```python
# This is for Google Drive (in my environment, may vary for your environment.)
filepath = f"/content/drive/MyDrive/machine_learning_challenge/ML-challenge/EX_challenging_problems/06_stock_price_expectation"

filename = input("Filename? (Blank if you want default): ")

if filename == None:
    pass
else:
    item_name = filename

specific_item_df.to_csv(f"{filepath}/{item_name}.csv", sep=',', encoding='utf-8')
specific_item_df.iloc[::-1].to_csv(f"{filepath}/{item_name}_reverse_order.csv", sep=',', encoding='utf-8')
print(f"Stored at {filepath}/{item_name}[/_reverse_order].csv")
```
