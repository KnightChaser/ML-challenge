```python
# HISTOGRAM (with normal distribution)

# Creating dataset
np.random.seed(0xCAFEABCD)
n_points = 100000
n_bins = 200

# Creating distribution
x = np.random.randn(n_points)
y = .8 ** x + np.random.randn(n_points) + 25

# Creating histogram
fig, axs = plt.subplots(1, 1,
						figsize = (8, 4),
						tight_layout = True)

axs.hist(x, bins = n_bins)
plt.title("Histogram with 100k random generations")
plt.xlabel("Distribution (sigma)")
plt.ylabel("Frequency")

plt.show()
```

```python
# SCATTER PLOT WITH RANDOM

random_x  = np.random.randint(1, 100 + 1, size = 100)
random_ya = np.random.randint(1, 100 + 1, size = 100)
random_yb = np.random.randint(1, 100 + 1, size = 100)

plt.scatter(random_x, random_ya, marker = "x")       # <--- Selectable marker
plt.scatter(random_x, random_yb, marker = "x")
plt.plot([0, 100], [0, 100], linestyle = "solid", color = "red")    # <-- Draw a line in the scatter plot

plt.title("Scatterplot with random dots")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["Random A", "Random B"])        # legend

plt.plot()
```

```python
# PIE CHART

platforms = ["Twitter", "Reddit", "Discord", "Bluesky", "Mastodon", "Instagram", "Facebook"]
data = [283, 110, 57, 32, 29, 11, 10]

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}% ({:d} hr)".format(pct, absolute)

fig = plt.figure(figsize = (10, 10))
plt.pie(data,
        autopct = lambda pct: func(pct, data),          # number marking
        labels = platforms,                             # platform classification
        shadow = True,                                  # shadow
        startangle = 20)                                # staring angle (turn the pie chart with the given angle value)
plt.title("My SNS usage")

plt.show()
```

```python
# 3D PLOT (Probably useful to visualize multivariable/3D functions)

# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
x = np.outer(np.linspace(-2, 2, 3200), np.ones(3200))   # "resolution" adjustment
y = x.copy().T                                          # transpose
z = (np.sin(x ** 2) + np.cos(y ** 2))

fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')

ax.plot_surface(x, y, z)

plt.title("z = sin²(x) + cos²(y)")
plt.show()
```

```python
# sin and cos

x_1 = np.arange(0, 2*np.pi, 0.1)
y_1 = np.sin(x_1)

x_2 = np.arange(0, 2*np.pi, 0.1)
y_2 = np.cos(x_2)

plt.plot(x_1, y_1, color = "green", label = "sine")
plt.plot(x_2, y_2, color = "magenta", label = "cosine")
plt.title("sine and cosine")
plt.legend()
plt.show()
```

```python
x = np.linspace(-1, 1, 100)

y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)  # create 2*2 figure

ax[0, 0].plot(x, y_1, label = "sin")
ax[0, 1].plot(x, y_2, label = "cos")
ax[1, 0].plot(x, y_3, label = "tan")
ax[1, 1].plot(x, y_4, label = "exp")

for _x in [0, 1]:
    for _y in [0, 1]:
        ax[_x, _y].legend()

plt.show()
```

### Draw Seaborn graph according to the given dataset

```python
df = pd.read_excel("https://github.com/KnightChaser/ML-challenge/raw/main/EX_note/health_2020.xlsx")        # Load excel file.
```

```python
df
```

```python
# With 2 variabales, draw a 2-dimensional graph

sns.lineplot(x = "waist", y = "weight", data = df)
```

```python
# Draw a 2-dimensional scatterplot with classificable data (hue)

sns.scatterplot(x = "systolic", y = "diastolic", data = df, hue = "smoking")
```

```python
# Grid plot (subplot, n*m grid)

g = sns.FacetGrid(df, col = "smoking", row = "drinking")
g.map(sns.scatterplot, "systolic", "diastolic")
```

```python
# General heatmap

data3 = df.loc[:,["gender", "drinking", "smoking"]]

plt.figure(figsize = (10, 6))
plt.title("3x3 Heat Map Graph(correlation)")

correlation_data3 = data3.corr()
sns.heatmap(correlation_data3, annot = True, cmap = "YlGnBu")
```

```python
# (More useful) heatmap

data8 = df.loc[:, ["gender", "weight", "waist", "drinking", "smoking", "cholesterol", "HDL", "LDL"]]

plt.figure(figsize = (13, 10))
plt.title("8x8 Heat Map Graph(correlation)")

correlation_data8 = data8.corr()
upp_mat = np.triu(correlation_data8)	# don't show upper (duplicated) triangle zone

sns.heatmap(correlation_data8, annot = True, cmap = "RdYlGn", mask = upp_mat)
```

```python
# Box plot
# Especially useful to find the outliers

plt.figure(figsize = (10, 6))
plt.title("Drinking & Weight Box Plot")

sns.boxplot(x = "drinking", y = "weight", hue = "gender", data = df)
plt.ylim(30, 100)
plt.show()
```

```python
# Facet grid
# Draw histogram according to the classifiable data, with distribution curve

fg = sns.FacetGrid(df, col = "drinking", height = 5, aspect = 1.3)
fg.map(sns.histplot, "waist", bins = 10, color = "g", kde = True)
```
