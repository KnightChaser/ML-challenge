```python
import numpy
```

```python
def softmax(input):
    exp_sum = sum(numpy.exp(input))
    result  = [numpy.exp(x) / exp_sum for x in input]
    return result

print(softmax([-3, -2, -1, 0, 1, 2, 3]))
# print(softmax([-1000, -500, 0, 500, 1000]))       # overflow
```

```python
def better_softmax(input):
    exp_sum = sum([numpy.exp(x - max(input)) for x in input])
    result  = [numpy.exp(x - max(input)) / exp_sum for x in input]
    return result

print(better_softmax([-3, -2, -1, 0, 1, 2, 3]))
print(better_softmax([-1000, -500, 0, 500, 1000]))       # OK!
```
