```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# BAR CHART (H)

years   = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
deposit = [208451, 211217, 249358, 388441, 410002, 444849, 498123]

plt.title("My deposit change")
plt.barh(years, deposit)
plt.ylabel("years")
plt.xlabel("money (USD Dollar)")

plt.show()
```

```python
# BAR CHART (Normal)

years   = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
deposit = [208451, 211217, 249358, 388441, 410002, 444849, 498123]

plt.title("My deposit change")
plt.bar(years, deposit)
plt.xlabel("years")
plt.ylabel("money (USD Dollar)")

plt.show()
```

```python
# STACKED BAR CHART

years  = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
assets = {
    "Cash"      : [172, 189, 190, 192, 207, 211, 284],
    "Crypto"    : [100, 108, 128, 131, 159, 108, 174],
    "Bond"      : [211, 240, 249, 251, 211, 199, 118],
    "Future"    : [41, 49, 50, 54, 70, 75, 90]
}

p1 = plt.bar(years, assets["Cash"], zorder=1)
p2 = plt.bar(years, assets["Crypto"], zorder=2)
p3 = plt.bar(years, assets["Bond"], zorder=3)
p4 = plt.bar(years, assets["Future"], zorder=4)

plt.title("My assets portfolio")
plt.ylabel("1,000 USD Dollar")
plt.xlabel("years")
plt.legend(assets.keys())
plt.show()
```

```python
city   = ['Delhi','Beijing','Washington','Tokyo','Moscow']
Gender = ['Male','Female']

pos = np.arange(len(city))

Happiness_Index_Male   = [60,40,70,65,85]
Happiness_Index_Female = [30,60,70,55,75]

plt.bar(pos,Happiness_Index_Male,color = 'blue',edgecolor = 'black')
plt.bar(pos,Happiness_Index_Female,color = 'pink',edgecolor = 'black',bottom = Happiness_Index_Male)
plt.xticks(pos, city)
plt.xlabel('City', fontsize = 16)
plt.ylabel('Happiness_Index', fontsize = 16)
plt.title('Stacked Barchart - Happiness index across cities',fontsize = 18)
plt.legend(Gender,loc = 2)
plt.show()
```
